```python
def find_c(n):
    #https://web.eecs.umich.edu/~cpeikert/pubs/lwe-analysis.pdf figure 1
    n2 = n   
    val = 0
    
    if(n2 == 128):
        val = 1.35
    if(n2 == 256):
        val = 1.25
    elif(n2 == 512):
        val = 1.72
    return val

def lemma31(n, q, sigma):

    c = find_c(n)
    lower_side = (sigma*sqrt(2*pi))^2
    t = floor(q/4)
    upper_frac = t*(sqrt(2)*pi)

    i=1
    while lower_side <= upper_frac/(c*(sqrt(2*n*ln(2/(10^(-i)))))):
        i = i+1
    
    return("10^", -(i-1))
```

```python
#failure rate medium security

p1_med_sec = lemma31(256, 4093, 3.33)
p2_med_sec = lemma31(256, 7681, 4.51)

print(p1_med_sec, p2_med_sec)
```

```python
#Failure Rate for 3.33, 3.05, 1.73

p1_1 = lemma31(256, 4093, 3.33)
p1_2 = lemma31(256, 4093, 3.05)
p1_3 = lemma31(256, 4093, 1.73)

p2_1 = lemma31(256, 7681, 4.49)
p2_2 = lemma31(256, 7681, 3.47)
p2_3 = lemma31(256, 7681, 1.77)

print(p1_1,p1_2,p1_3)
print(p2_1,p2_2,p2_3)
```
